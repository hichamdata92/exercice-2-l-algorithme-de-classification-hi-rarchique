```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from scipy.spatial.distance import pdist, squareform
import pandas as pd

def hierarchical_clustering_analysis():
    """
    Analyse complète de classification hiérarchique - Exercice 2
    Implémente les méthodes de lien maximum et Ward
    """
    
    # =====================================
    # DONNÉES D'ENTRÉE
    # =====================================
    # Remplacez ces données par celles de votre exercice 1
    data = np.array([
        [1, 2],  # Point A
        [3, 1],  # Point B
        [2, 4],  # Point C
        [5, 3],  # Point D
        [4, 5]   # Point E
    ])
    labels = ['A', 'B', 'C', 'D', 'E']
    
    print("=" * 60)
    print("CLASSIFICATION HIÉRARCHIQUE - EXERCICE 2")
    print("=" * 60)
    
    print("\nDonnées d'entrée:")
    df_data = pd.DataFrame(data, index=labels, columns=['x', 'y'])
    print(df_data)
    
    # =====================================
    # 1. MATRICE DES DISTANCES
    # =====================================
    print("\n" + "=" * 50)
    print("1. MATRICE DES DISTANCES EUCLIDIENNES")
    print("=" * 50)
    
    distances = pdist(data, metric='euclidean')
    distance_matrix = squareform(distances)
    
    df_distances = pd.DataFrame(distance_matrix, index=labels, columns=labels)
    print("\nMatrice des distances:")
    print(df_distances.round(3))
    
    # =====================================
    # 2. CLASSIFICATION AVEC LIEN MAXIMUM
    # =====================================
    print("\n" + "=" * 50)
    print("2. CLASSIFICATION AVEC LIEN MAXIMUM")
    print("=" * 50)
    
    # Calcul de la hiérarchie
    linkage_max = linkage(data, method='complete')
    
    print("\nHiérarchie indicée (method='complete'):")
    print("-" * 60)
    print("Étape | Cluster1 | Cluster2 | Distance | Nb points | Nouveau cluster")
    print("-" * 60)
    
    cluster_names = list(labels)  # Copie pour tracking des noms
    
    for i, merge in enumerate(linkage_max):
        idx1, idx2 = int(merge[0]), int(merge[1])
        
        # Détermination des noms de clusters
        if idx1 < len(labels):
            name1 = labels[idx1]
        else:
            name1 = f"C{idx1}"
            
        if idx2 < len(labels):
            name2 = labels[idx2]
        else:
            name2 = f"C{idx2}"
        
        new_cluster = f"({name1},{name2})"
        
        print(f"{i+1:5d} | {name1:8s} | {name2:8s} | {merge[2]:8.3f} | {int(merge[3]):9d} | {new_cluster}")
    
    # Partition en 2 classes
    clusters_max_2 = fcluster(linkage_max, 2, criterion='maxclust')
    print(f"\nPartition en 2 classes (Lien Maximum):")
    print("-" * 35)
    for i, label in enumerate(labels):
        print(f"Point {label} -> Classe {clusters_max_2[i]}")
    
    # Analyse des classes
    print(f"\nComposition des classes:")
    for classe in range(1, 3):
        points_classe = [labels[i] for i, c in enumerate(clusters_max_2) if c == classe]
        print(f"Classe {classe}: {{{', '.join(points_classe)}}}")
    
    # =====================================
    # 3. CLASSIFICATION AVEC WARD
    # =====================================
    print("\n" + "=" * 50)
    print("3. CLASSIFICATION AVEC CRITÈRE DE WARD")
    print("=" * 50)
    
    # Calcul de la hiérarchie avec Ward
    linkage_ward = linkage(data, method='ward')
    
    print("\nHiérarchie indicée (method='ward'):")
    print("-" * 60)
    print("Étape | Cluster1 | Cluster2 | Inertie  | Nb points | Nouveau cluster")
    print("-" * 60)
    
    for i, merge in enumerate(linkage_ward):
        idx1, idx2 = int(merge[0]), int(merge[1])
        
        # Détermination des noms de clusters
        if idx1 < len(labels):
            name1 = labels[idx1]
        else:
            name1 = f"C{idx1}"
            
        if idx2 < len(labels):
            name2 = labels[idx2]
        else:
            name2 = f"C{idx2}"
        
        new_cluster = f"({name1},{name2})"
        
        print(f"{i+1:5d} | {name1:8s} | {name2:8s} | {merge[2]:8.3f} | {int(merge[3]):9d} | {new_cluster}")
    
    # Partition en 2 classes
    clusters_ward_2 = fcluster(linkage_ward, 2, criterion='maxclust')
    print(f"\nPartition en 2 classes (Ward):")
    print("-" * 30)
    for i, label in enumerate(labels):
        print(f"Point {label} -> Classe {clusters_ward_2[i]}")
    
    # Analyse des classes
    print(f"\nComposition des classes:")
    for classe in range(1, 3):
        points_classe = [labels[i] for i, c in enumerate(clusters_ward_2) if c == classe]
        print(f"Classe {classe}: {{{', '.join(points_classe)}}}")
    
    # =====================================
    # 4. COMPARAISON DES MÉTHODES
    # =====================================
    print("\n" + "=" * 50)
    print("4. COMPARAISON DES DEUX MÉTHODES")
    print("=" * 50)
    
    print("\nPartitions en 2 classes:")
    print("-" * 30)
    print("Point | Lien Max | Ward")
    print("-" * 30)
    for i, label in enumerate(labels):
        print(f"  {label}   |    {clusters_max_2[i]}     |  {clusters_ward_2[i]}")
    
    # Calcul de l'accord entre les deux méthodes
    accord = sum(1 for i in range(len(labels)) if clusters_max_2[i] == clusters_ward_2[i])
    print(f"\nAccord entre les méthodes: {accord}/{len(labels)} points ({100*accord/len(labels):.1f}%)")
    
    # =====================================
    # 5. VISUALISATION
    # =====================================
    print("\n" + "=" * 50)
    print("5. GÉNÉRATION DES GRAPHIQUES")
    print("=" * 50)
    
    # Configuration de la figure
    fig = plt.figure(figsize=(16, 12))
    
    # Sous-graphique 1: Données originales
    ax1 = plt.subplot(2, 3, 1)
    scatter = ax1.scatter(data[:, 0], data[:, 1], s=150, c='blue', alpha=0.7, edgecolors='black')
    for i, label in enumerate(labels):
        ax1.annotate(label, (data[i, 0], data[i, 1]), 
                    xytext=(8, 8), textcoords='offset points',
                    fontsize=12, fontweight='bold')
    ax1.set_title('Données originales', fontsize=14, fontweight='bold')
    ax1.set_xlabel('x', fontsize=12)
    ax1.set_ylabel('y', fontsize=12)
    ax1.grid(True, alpha=0.3)
    
    # Sous-graphique 2: Dendrogramme Lien Maximum
    ax2 = plt.subplot(2, 3, 2)
    dend1 = dendrogram(linkage_max, labels=labels, ax=ax2, 
                      orientation='top', distance_sort='descending',
                      show_leaf_counts=True)
    ax2.set_title('Dendrogramme - Lien Maximum', fontsize=14, fontweight='bold')
    ax2.set_ylabel('Distance', fontsize=12)
    
    # Sous-graphique 3: Dendrogramme Ward
    ax3 = plt.subplot(2, 3, 3)
    dend2 = dendrogram(linkage_ward, labels=labels, ax=ax3, 
                      orientation='top', distance_sort='descending',
                      show_leaf_counts=True)
    ax3.set_title('Dendrogramme - Ward', fontsize=14, fontweight='bold')
    ax3.set_ylabel('Inertie', fontsize=12)
    
    # Sous-graphique 4: Partition Lien Maximum
    ax4 = plt.subplot(2, 3, 4)
    colors = ['red', 'blue', 'green', 'orange', 'purple']
    for classe in range(1, 3):
        mask = clusters_max_2 == classe
        ax4.scatter(data[mask, 0], data[mask, 1], 
                   c=colors[classe-1], s=150, alpha=0.7,
                   label=f'Classe {classe}', edgecolors='black')
    
    for i, label in enumerate(labels):
        ax4.annotate(f"{label}", (data[i, 0], data[i, 1]), 
                    xytext=(8, 8), textcoords='offset points',
                    fontsize=12, fontweight='bold', color='white')
    
    ax4.set_title('Partition - Lien Maximum', fontsize=14, fontweight='bold')
    ax4.set_xlabel('x', fontsize=12)
    ax4.set_ylabel('y', fontsize=12)
    ax4.legend()
    ax4.grid(True, alpha=0.3)
    
    # Sous-graphique 5: Partition Ward
    ax5 = plt.subplot(2, 3, 5)
    for classe in range(1, 3):
        mask = clusters_ward_2 == classe
        ax5.scatter(data[mask, 0], data[mask, 1], 
                   c=colors[classe-1], s=150, alpha=0.7,
                   label=f'Classe {classe}', edgecolors='black')
    
    for i, label in enumerate(labels):
        ax5.annotate(f"{label}", (data[i, 0], data[i, 1]), 
                    xytext=(8, 8), textcoords='offset points',
                    fontsize=12, fontweight='bold', color='white')
    
    ax5.set_title('Partition - Ward', fontsize=14, fontweight='bold')
    ax5.set_xlabel('x', fontsize=12)
    ax5.set_ylabel('y', fontsize=12)
    ax5.legend()
    ax5.grid(True, alpha=0.3)
    
    # Sous-graphique 6: Analyse des distances
    ax6 = plt.subplot(2, 3, 6)
    
    # Graphique des hauteurs de fusion
    steps = range(1, len(linkage_max) + 1)
    heights_max = linkage_max[:, 2]
    heights_ward = linkage_ward[:, 2]
    
    ax6.plot(steps, heights_max, 'o-', label='Lien Maximum', linewidth=2, markersize=8)
    ax6.plot(steps, heights_ward, 's-', label='Ward', linewidth=2, markersize=8)
    ax6.set_title('Hauteurs de fusion', fontsize=14, fontweight='bold')
    ax6.set_xlabel('Étape de fusion', fontsize=12)
    ax6.set_ylabel('Hauteur', fontsize=12)
    ax6.legend()
    ax6.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print("Graphiques générés avec succès!")
    
    # =====================================
    # 6. EXPORT DES RÉSULTATS
    # =====================================
    print("\n" + "=" * 50)
    print("6. RÉSULTATS POUR VÉRIFICATION MANUELLE")
    print("=" * 50)
    
    print(f"\nMatrice des distances (format pour LaTeX):")
    print("\\begin{tabular}{|c|" + "c|" * len(labels) + "}")
    print("\\hline")
    header = " & " + " & ".join(labels) + " \\\\"
    print(header)
    print("\\hline")
    for i, label in enumerate(labels):
        row = label + " & " + " & ".join([f"{distance_matrix[i,j]:.2f}" for j in range(len(labels))]) + " \\\\"
        print(row)
    print("\\hline")
    print("\\end{tabular}")
    
    return {
        'data': data,
        'labels': labels,
        'distance_matrix': distance_matrix,
        'linkage_max': linkage_max,
        'linkage_ward': linkage_ward,
        'clusters_max_2': clusters_max_2,
        'clusters_ward_2': clusters_ward_2
    }

def manual_max_linkage_step_by_step(data, labels):
    """
    Implémentation manuelle du lien maximum pour vérification
    """
    print("\n" + "="*50)
    print("VÉRIFICATION MANUELLE - LIEN MAXIMUM")
    print("="*50)
    
    # Initialisation
    clusters = [[i] for i in range(len(data))]
    cluster_names = labels.copy()
    distances = pdist(data)
    dist_matrix = squareform(distances)
    
    step = 0
    while len(clusters) > 1:
        step += 1
        print(f"\n--- ÉTAPE {step} ---")
        print(f"Clusters actuels: {cluster_names}")
        
        # Trouver la paire avec distance minimum
        min_dist = float('inf')
        best_i, best_j = -1, -1
        
        for i in range(len(clusters)):
            for j in range(i+1, len(clusters)):
                # Calculer distance lien maximum entre clusters i et j
                max_dist = 0
                for point1 in clusters[i]:
                    for point2 in clusters[j]:
                        max_dist = max(max_dist, dist_matrix[point1, point2])
                
                print(f"Distance({cluster_names[i]}, {cluster_names[j]}) = {max_dist:.3f}")
                
                if max_dist < min_dist:
                    min_dist = max_dist
                    best_i, best_j = i, j
        
        print(f"Fusion: {cluster_names[best_i]} ∪ {cluster_names[best_j]} (distance = {min_dist:.3f})")
        
        # Fusionner les clusters
        new_cluster = clusters[best_i] + clusters[best_j]
        new_name = f"({cluster_names[best_i]},{cluster_names[best_j]})"
        
        # Supprimer les anciens clusters (en commençant par le plus grand index)
        if best_j > best_i:
            clusters.pop(best_j)
            clusters.pop(best_i)
            cluster_names.pop(best_j)
            cluster_names.pop(best_i)
        else:
            clusters.pop(best_i)
            clusters.pop(best_j)
            cluster_names.pop(best_i)
            cluster_names.pop(best_j)
        
        # Ajouter le nouveau cluster
        clusters.append(new_cluster)
        cluster_names.append(new_name)

if __name__ == "__main__":
    # Exécution principale
    results = hierarchical_clustering_analysis()
    
    # Vérification manuelle optionnelle
    print("\n" + "="*60)
    response = input("Voulez-vous voir la vérification manuelle du lien maximum? (o/n): ")
    if response.lower() in ['o', 'oui', 'y', 'yes']:
        manual_max_linkage_step_by_step(results['data'], results['labels'])
    
    print("\n" + "="*60)
    print("ANALYSE TERMINÉE")
    print("="*60)
```
